Agora, vamos trabalhar com os dados de quantidade de falecidos e recuperados da COVID-19 (que foram carregados anteriormente)

```python
display(deaths.head())
display(recovered.head())
```

```python
confirmed.groupby("Country/Region").sum()[last_date].head()
```

```python
def latest_by_country(data):
  return data.groupby("Country/Region").sum().iloc[:,-1]

latest_by_country(confirmed).head()
```

```python
information = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]

combined = pd.concat(information, axis=1)
```

```python
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

```python
display(latest_by_country(confirmed).head())
display(latest_by_country(deaths).head())
display(latest_by_country(recovered).head())
```

Dos dados brutos, pode-se calcular a taxa de mortalidade em percentagem (%) = nÃºmero de casos letais dentro dos casos letais e nÃ£o letais

```python
sum_up = combined.sum()
display(sum_up)
```

```python
combined
```

```python
lethality_rate_1 = sum_up["deaths"] / sum_up["confirmed"] * 100
lethality_rate_1
print("lethality_rate_1 = " + str(lethality_rate_1))

lethality_rate_2 = sum_up["deaths"] / (sum_up["deaths"] + sum_up["recovered"]) * 100 #Logicamente, isso deveria fazer sentido
print("lethality_rate_2 (nÃ£o confiÃ¡vel) = " + str(lethality_rate_2) + " (OBS.: Por algum motivo, o nÃºmero de pacientes recuperados ficou igual a zero no DataFrame)")
```

```python
print(lethality_rate_1 * 100)
```

```python
combined
```

```python
#Calcular taxa de letalidade por paÃ­s
lethality_rate_1 = (combined["deaths"] / combined["confirmed"] * 100)
lethality_rate_2 = (combined["deaths"] / (combined["deaths"] + combined["recovered"]) * 100)
combined["lethality_rate_1"] = lethality_rate_1
combined["lethality_rate_2_invalid_data"] = lethality_rate_2

combined.head()
```

```python
combined.sort_values("lethality_rate_1", ascending=False).head(10)
```

```python
combined[["confirmed", "lethality_rate_1"]]
```

```python
plt.figure(figsize=(20,5))
plt.title("Casos confirmados versus taxa de letalidade")
plt.plot(combined[["confirmed", "lethality_rate_1"]])# NÃ£o faz sentido plotar nÃºmero de casos confirmados (nÃºmeros inteiros) e
plt.legend(["confirmed", "lethality_rate_1"])        # a taxa de letalidade (percentagem) no mesmo eixo, pois estÃ£o em escalas diferentes
plt.show()# exemplo de pÃ©ssima visualizaÃ§Ã£o
```

```python
import seaborn as sns
# mostrar no grÃ¡fico de dispersÃ£o apenas os paÃ­ses que possuem mais de 40 casos confirmados (taxa de letalidade mais prÃ³xima do real)
sns.scatterplot(data = combined.query("confirmed > 40"), x = "confirmed", y = "lethality_rate_1")
```

```python
sns.distplot(combined.query("confirmed > 40")["lethality_rate_1"], kde=False)
```
